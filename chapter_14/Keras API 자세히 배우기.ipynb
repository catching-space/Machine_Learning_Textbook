## Import

````python
import tensorflow as tf
````

## 랜덤 데이터 만들기

````python
np.random.seed(190917)
````

````python
def make_random_data():
    x = np.random.uniform(low=-2, high=2, size=200)
    y = []
    for t in x:
        r = np.random.normal(loc=0.0,
                             scale=(0.5 + t*t/3),
                             size=None)
        y.append(r)
    return x, 1.726*x - 0.84 + np.array(y)
````

````python
x, y = make_random_data()
plt.plot(x, y, 'o')
plt.show()
````

````python
X_train, y_train = x[:150], y[:150]
X_test, y_test = x[150:], y[150:]
````

## tf.keras API로 초간단한 회귀분석 신경망 모델링하기

````python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(units=1, input_dim=1))
model.summary()
````

````python
model.compile(optimizer='sgd', loss='mse')
history = model.fit(X_train, y_train, epochs=500,
                    validation_split=0.3)
````

````python
epochs = np.arange(1, 500+1)
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
````

## Keras 함수형 API 활용하기

````python
from tensorflow.keras import Model, Input
````

````python
input = tf.keras.Input(shape=(1,))
output = tf.keras.layers.Dense(1)(input)
````

Dense 클래스를 마치 함수처럼 호출하고 있다. 클래스 객체를 함수처럼 호출할 수 있는 파이썬의 특수한 `__call__()` 메서드를 사용한 것. 두 번째 코드는 다음과 같이 쓸 수 있다.

```
dense = tf.keras.layers.Dense(1)
output = dense(input)
```

혹은

```
dense = tf.keras.layers.Dense(1)
output = dense.__call__(input)
```

Dense 층의 객체(dense)를 네트워크의 다른 부분에 재사용할 일이 없다면(즉, 이 층의 가중치를 공유하지 않는다면) 객체 생성과 동시에 호출하는 것이 편리하다. 그 다음 Model 클래스에 input과 output을 전달하여 신경망 모델을 만들어보자.

````python
model = tf.keras.Model(input, output)
model.summary()
````

````python
model.compile(optimizer='sgd', loss='mse')
history = model.fit(X_train, y_train, epochs=500,
                    validation_split=0.3)
````

````python
epochs = np.arange(1, 500+1)
plt.plot(epochs, history.history['loss'], label='Training loss')
plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
````

## tf.keras 모델의 저장과 복원

#### 가중치 저장

````python
model.save_weights('simple_weights.h5')
````

#### 가중치 로드

````python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=1))
model.compile(optimizer='sgd', loss='mse')
model.load_weights('simple_weights.h5')
````

가중치가 제대로 로드되었는지 확인하기 위해 테스트셋에 대한 loss값을 확인해보자

````python
model.evaluate(X_test, y_test)
````

가중치 외에 모델 전체를 저장하는 것도 가능하다.

````python
model.save('simple_model.h5')
````

````python
model = tf.keras.models.load_model('simple_model.h5')
model.evaluate(X_test, y_test)
````

## `ModelCheckpoint`, `EarlyStopping` callback

모델을 훈련하는 동안 ModelCheckpoint callback을 사용하여 최고의 성능을 내는 가중치를 저장할 수 있다. 보통 더 이상 성능이 개선되지 않을 때 훈련을 멈추도록 하는 EarlyStopping callback과 함께 사용하는 경우가 많다.

````python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=1))
model.compile(optimizer='sgd', loss='mse')

callback_list = [tf.keras.callbacks.ModelCheckpoint(filepath='my_model.h5',
                                                   monitor='val_loss', save_best_only=True),
                 tf.keras.callbacks.EarlyStopping(patience=5)] # 5 epoch 동안 개선되지 않으면 훈련 중단

history = model.fit(X_train, y_train, epochs=500,
                    validation_split=0.2, callbacks=callback_list)
````

아까 저장한 모델을 로드한 후 ModelCheckpoint callback에서 저장한 가중치를 load_weights() 메서드로 적재해보자.

````python
model = tf.keras.models.load_model('simple_model.h5')
model.load_weights('my_model.h5')
model.evaluate(X_test, y_test)
````
